### b) (3,0) Apresente o vetor média e a matriz covariância.

```python
df_var = df[['X1: Vitórias', 'X2: Gols Pró']]
matriz = df_var.to_numpy()
print(matriz.shape)
print(matriz.mean(axis=0))
vetor_media = df_var.mean()
matriz_cov = df_var.cov()

print(f'Vetor Média: {vetor_media}')
print(50*'-')
print(f'Matriz covariância: {matriz_cov}')
```

### c) (2,0) Apresente a matriz correlação entre as duas variáveis e interprete.

```python
matriz_corr = df_var.corr()
sns.heatmap(matriz_corr, annot=True)
plt.show()
```
