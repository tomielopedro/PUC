**5. O dataset possui qualidade para uso em Machine Learning?**

**Parcialmente.** Há pelo menos dois problemas evidentes já nas 5 primeiras linhas:

**Problema 1 — Valor inválido em `precip` (linha 0):**
O valor `-1.0` para precipitação é fisicamente impossível — precipitação não pode ser negativa. Provavelmente representa um valor ausente codificado numericamente (convenção comum no INMET: `-1` ou `-9999` para dados faltantes). Isso precisa ser tratado antes de qualquer modelagem.

**Problema 2 — Valores absurdos em `vel_vento`:**
Os valores `2333333.0`, `1333333.0` e `3666667.0` são claramente erros de dados — velocidade do vento medida em m/s ou km/h jamais atingiria esses valores. Indica provável erro de parsing (vírgula interpretada como separador decimal incorreto, ou encoding quebrado). A linha 2 tem `2.0`, que seria o único valor plausível.

**Resumo dos problemas:**

| Problema | Coluna afetada | Ação recomendada |
|---|---|---|
| Valor negativo impossível | `precip` | Tratar como NaN e imputar ou remover |
| Valores absurdos | `vel_vento` | Investigar encoding/parsing, corrigir ou remover |
| Possíveis outros missing values | Todas | Verificar toda a série com `df.isnull()` e `df.describe()` |
